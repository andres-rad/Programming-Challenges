```python
import pygraphviz
import networkx as nx
```

```python
G = nx.Graph()
for x in range(59):
  G.add_node(x+1)
```

```python
s = '21 40 59 11 59 28 9 55 29 26 59 47 30 1 47 58 59 55 59 47 36 12 59 11 15 1 1 34 1 30 48 23 28 24 20 42 46 48 58 49 26 9 29 35 10 18 15 58 43 18 56 42 6 20 55 5 15 26'
l = list(map(int, s.split()))
```

```python
for x in range(len(l)):
#   print("{} {}".format(x+2, l[x]))
  G.add_edge(l[x], x+2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
from networkx.drawing.nx_agraph import write_dot, graphviz_layout

plt.title('draw_networkx')
# plt.set_size_inches(10, 10)

pos =graphviz_layout(G, prog='dot')
nx.draw(G, pos, with_labels=True, arrows=True)
plt.savefig('nx_test.png')
plt.show()
```
